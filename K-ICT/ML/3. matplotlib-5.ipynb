### plt.imshow()를 이용한 이미지 플롯

```python
from sklearn.datasets import load_digits
```

- sklearn datasets의 load_digits 함수로 손글씨 데이터 로딩

```python
digits = load_digits()
digits.keys()
```

```python
type(digits)
```

- digits의 타입은 Bunch

- Bunch라는 객체는 keys 함수 제공

```python
digits.images[0]
```

- 이미지 정보는 픽셀 정보(8x8)로 구성되어 있음

- 손글씨 이미지는 밝기 정보를 이용해 표현해 놓았음
    - 흑백
    
- 메모리만 차지하는 데이터는 희소행렬을 이용해 변환해야 함

```python
# 디지털 이미지 그리기
fig, ax = plt.subplots(10, 10, figsize=(5, 5))
for i, ax_i in enumerate(ax.flat):
    ax_i.imshow(digits.images[i], cmap="binary")
    ax_i.set(xticks=[], yticks=[])
```

- subplots()
    - 10행 10열 구성

- ax.flat
    - Axes를 flat 시키기
    - 2차원 구조 ==> 1차원 구조

- enumerate 함수를 이용해 Figure 객체와 Axes 얻기

- binary: 흑백

- ax_i.set()
    - x, y축 정보를 기술하지 않고 공간만 구성(빈 리스트)
        - 공간만 존재
        - 가로축, 세로축 정보는 기술 하지 않게 하려고... 


- 필기 인식이 얼마큼 잘 되는지 라벨을 주어 그래픽 결과를 만들어 내는 데에 이미지 플롯이 유용함
