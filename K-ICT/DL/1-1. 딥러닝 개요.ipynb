# 1-1. 전체 교육 과정 소개

### 1. 데이터 기반 학습

- 학습이라는 추상적인 개념의 구체화

- 학습 개념을 수학적으로 이해하기

- 데이터 기반 학습에 대한 이해하기

<br/>

### 2. 파이썬 고속 데이터 처리

- 다차원, 고속 배열 지원

- 다차원 배열에 대한 직접 연산 지원

- 다양한 선형 대수 함수 지원

- 행렬 변환 함수 지원

- 신회성 있고 검증된 수치 연산 함수 라이브러리

- 기본적인 numpy 모듈의 이해와 수학점수 및 주가지수 예측 실습

<br/>

### 3. TensorFlow 데이터 연산

- TensorFlow의 기본개념과 원리

- tensor 정의 및 사칙연산 수행

- TensorFlow를 이용한 다양한 코드 실습

<br/>

### 4. 주가지수 예측 및 데이터 시각화

- 다양한 형태의 텐서 행렬 생성

- 평균 최댓값을 쉽게 계산할 수 있는 reduce 함수

- 텐서 간 행렬 및 역행렬

- 수학점수 예측 및 주가지수 예측 및 데이터 시각화 실습

<br/>

### 5. 선형회귀: 집값 예측

- 데이터(직선으로 가정)구성, 여러 데이터 구성 요인과 결과의 관련성 확인을 위한 선형회귀

- 특성 요인에 따라 결과가 어떻게 변동되는지 예측 가능

- 신경망 학습 알고리즘의 핵심인 경사하강법 이해 및 실습

- 경사하강법을 개선할 수 있는 TensorFlow 함수 및 보스톤 교외 주택 집값 예측 학습모델 실습

<br/>

### 6. 로지스틱 회귀: 스팸 검출기

- 전체 데이터를 2개의 그룹으로 분류하여 로지스틱 회귀

- 선형회귀와 로지스틱 회귀의 공통점과 차이점

- TensorFlow를 사용해 로지스틱 회귀 학습 구현 실습

- 당뇨병 예측 및 스팸 필터링 예제 실습

<br/>

### 7. 다중분류: 와인 품질 및 붓꽃 품종 분류

- 여러 범주의 데이터를 분류할 수 있는 SoftMax라는 학습 알고리즘

- 붓꽃 품종 분류와 와인 분류 예제 실습

<br/>

### 8. 인공 신경망의 이해

- 신경망을 기반으로 만들어진 인공신경망에 대한 이해

- 인간의 신경망의 메커니즘을 단순화한 단순 퍼셉트론 이해

- NOT, AND, OR 등의 기본 논리 소자를 퍼셈트론으로 교환

- 단순 퍼셉트론의 한계 이해

<br/>

### 9. 다층신경망: 암진단 예측

- 은닉층 추가로 구조를 변경한 다층퍼셉트론

- 다층구조를 가지고 있는 신경 학습

- 암 진단 데이터를 사용한 다층 퍼셉트론 실습

<br/>

### 10. 사람의 시각을 닮은 신경망

- 인간의 시각적 특성을 반영한 컨볼루션 신경망

- 공간 데이터 모델링

- 2D 영상 필터링 실습

<br/>

### 11. 컨볼루션 신경망: 도형 인식

- 물체의 이동, 회전, 크기의 변형 특성을 반영한 컨볼루션 신경망 개념

- 컨볼루션 신경망의 계층 구성과 각 계층의 특징 이해

- 도형인식기 및 손글씨 인식기 실습

<br/>

### 12. 깊은 신경망 딥러닝 최적화

- 모멘텀(Momentum)

    - 진행중인 방향으로 관성을 더해 지역 최소점(local minimum)에 빠지지 않도록 함

- ADAM(Adaptive Moment Estimation)

    - 기울기의 지수 평균과 기울기의 제곱의 지수 평균

<br/>

### 13. 딥러닝을 이용한 영상 인식

- 손글씨 인식기 및 패션 영상 인식기 구현

- 딥러닝 최적화 - 모멘텀

- 딥러닝 최적화 - 드롭아웃(dropout)
