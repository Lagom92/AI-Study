# 4. 히스토그램

```python
import matplotlib.pyplot as plt
import numpy as np
from seaborn import load_dataset
```

### 데이터의 분포 형상을 식별하는 히스토그램

```python
np.random.seed(0)

# 평균
mu = 72
# 표준편차
sigma = 8

x = np.random.normal(mu, sigma, 10000)
plt.hist(x, rwidth=0.9, bins=16)
```

- bins=16
    - 전체 계급 구간이 16이라는 의미

- rwidth=0.9
    - 히스토그램 전체 크기(1)의 0.9

### tips 데이터를 활용해 히스토그램 그리기

```python
tips = load_dataset("tips")
tips.info()
```

```python
tips.head()
```

```python
plt.hist(x=tips.tip, rwidth=0.9, bins=20)
plt.title("tips")
plt.xlabel("tip")
plt.ylabel("frequency")
```

- 계급구간(20개) 얼마나 많은 빈도수(ylabel)가 있는가?

### 상대도수 히스토그램과 확률밀도 함수

```python
import scipy as sp
```

```python
bins = 20
plt.hist(x=tips.tip, rwidth=0.9, bins=bins, density=True)
plt.title("tips")
plt.xlabel("tip")
plt.ylabel("density")

mu, sigma = tips.tip.mean(), tips.tip.std()
tip_min = tips.tip.min()
tip_max = tips.tip.max()
x = np.linspace(tip_min, tip_max, bins)
y = sp.stats.norm(mu, sigma).pdf(x)
plt.plot(x, y)
```

- ylabel을 density로 표현하여 정규분포의 근사치 얻기
- 확률 밀도 함수와 관련된 내용을 연결해 볼 수 있음

- 평균 및 표준편차 값
    - tip의 mean, tip의 std(standard deviation) 필요

- min, max
    - x와 y의 경계값 지정

- scipy에서 normal distribution의 확률밀도함수를 활용해 y값 얻기
    - sp.stats.norm()
    - pdf() : 확률밀도함수
    - 평균과 표준편차를 전달해 정규분포모델을 만들어 y값 얻기
