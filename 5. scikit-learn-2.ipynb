## Scikit-Learn Estimator API 기본 활용 절차

1. 데이터 준비

2. 모델 클래스 선택

3. 모델 인스턴스 생성과 하이퍼파라미터 선택

4. 특징 행렬과 대상 벡터 준비

5. 모델을 데이터에 적합(Fit)

6. 새로운 데이터를 이용해 예측

7. 모델 평가(정확도)


- 추정기

### 1. 데이터 준비

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rs = np.random.RandomState(10)
x = 10 * rs.rand(100)
y = 3 * x + 2 * rs.rand(100)
```

```python
plt.scatter(x, y, s=10)
```

- 선형적인 데이터

### 2. 모델 클래스 선택

### 3. 모델 인스턴스 생성과 하이퍼파라미터 선택

```python
from sklearn.linear_model import LinearRegression

regr = LinearRegression()

# 하이퍼파라미터를 설정 하고자 할 경우
# regr = LinearRegression(fit_intercept=True)
```

- 선형회귀 모델

### 4. 특징 행렬과 대상 벡터 준비

```python
X = x.reshape(-1, 1)

X.shape, y.shape
```

### 5. 모델을 데이터에 적합

```python
regr.fit(X, y)
```

```python
regr.coef_
```

- coef_

    - 직선의 기울기

```python
regr.intercept_
```

- intercept_

    - y 절편

### 6. 새로운 데이터를 이용해 예측

```python
x_new = np.linspace(-1, 11, num=100)
```

```python
X_new = x_new.reshape(-1, 1)

X_new.shape
```

```python
y_pred = regr.predict(X_new)
```

```python
plt.plot(x_new, y_pred, c="red")
plt.scatter(x, y, s=10)
```

- 예측선이 만들어짐(빨간색)

### 7. 모델 평가

```python
# RMSE
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y, y_pred))
```

```python
np.round(rmse, 2)
```

- error는 0에 가까울 수록 좋음
